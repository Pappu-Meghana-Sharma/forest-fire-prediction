```python
import numpy as np
from scipy.stats import chi2,shapiro

X_outliers = df_combined[['Temperature', 'RH', 'Ws', 'Rain', 'FFMC', 'DMC', 'DC', 'ISI', 'BUI']].dropna()

mean = np.mean(X_outliers, axis=0)
cov = np.cov(X_outliers, rowvar=False)
inv_cov = np.linalg.inv(cov)

def mahalanobis(x):
    diff = x - mean
    return np.sqrt(diff.T @ inv_cov @ diff)

md = X_outliers.apply(lambda row: mahalanobis(row.values), axis=1)

# Chi-Square threshold for 9 continuous variables at p < 0.05
threshold = chi2.ppf(0.95, df=X_outliers.shape[1])
outliers_count = (md > threshold).sum()

print(f"Multivariate Outliers detected (Mahalanobis > {threshold:.2f}): {outliers_count}")
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stats
sorted_md = np.sort(md)
chi_sq = stats.chi2.ppf(np.linspace(0.01, 0.99, len(sorted_md)), df=X_outliers.shape[1])

plt.scatter(chi_sq, sorted_md)
plt.plot(chi_sq, chi_sq, color='red')
plt.xlabel("Theoretical Chi-Square Quantiles")
plt.ylabel("Observed Mahalanobis Distance")
plt.title("MVN Check using Mahalanobis Q-Q Plot")
plt.show()
```

Since the points does not lie on the line, it deviates from Multivariate Normality

```python
# ---  REGRESSION & MULTICOLLINEARITY (VIF) ---
from statsmodels.stats.outliers_influence import variance_inflation_factor

#  Formal Multicollinearity Check
X_vif = df_combined[['Temperature', 'RH', 'Ws', 'Rain', 'FFMC', 'DMC', 'DC', 'ISI', 'BUI']]
vif_data = pd.DataFrame({"feature": X_vif.columns,
                         "VIF": [variance_inflation_factor(X_vif.values, i) for i in range(len(X_vif.columns))]})
print("Variance Inflation Factors (VIF > 10 is problematic):")
display(vif_data.sort_values(by="VIF", ascending=False))
```

Most features have extremely high VIF values (>10), indicating severe multicollinearity.

Fire-related indices (BUI, DMC, FFMC, DC, etc.) are highly interdependent and derived from similar factors.

Only Rain (VIF ≈ 1.6) behaves as an independent feature.

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
#  Linear Regression (Target: FWI)
X_train, X_test, y_train, y_test = train_test_split(X_vif, df_combined['FWI'], test_size=0.2, random_state=42)
reg_model = LinearRegression().fit(X_train, y_train)
y_pred = reg_model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
#  Diagnostics (Adequacy Testing)
residuals = y_test - y_pred
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
ax1.scatter(y_pred, residuals, alpha=0.5); ax1.axhline(0, color='red'); ax1.set_title("Residual Plot")
sns.histplot(residuals, kde=True, ax=ax2); ax2.set_title("Residual Distribution (Normality)")
plt.show()

print(f"Initial Regression R2 Score: {r2_score(y_test, y_pred):.4f}")
```

**Residual Plot:**
Residuals are mostly centered around zero, which is good.
However, a slight pattern/uneven spread is visible → indicates minor non-linearity or heteroscedasticity.

**Residual Distribution :**
Residuals are approximately normally distributed, but not perfectly symmetric.
Slight skewness suggests small deviations from normality.

The model achieved a very high R² (≈ 0.9883), explaining ~98.8% of FWI variance, indicating strong predictive performance. This is expected since FWI is derived from the input features, meaning the model primarily reconstructs it. The high multicollinearity among features contributes to this high accuracy

```python
from factor_analyzer import FactorAnalyzer

X_vif = df_combined.drop(columns=['Classes', 'Region', 'FFMC_Danger_Level', 'FFMC_Danger_Numeric', 'month'], errors='ignore')

fa = FactorAnalyzer(n_factors=3, rotation='varimax')
fa.fit(X_scaled)

# Now shapes will match: (10, 3)
loadings_fa = pd.DataFrame(
    fa.loadings_,
    index=X_vif.columns,
    columns=['Dryness Factor', 'Atmospheric Factor', 'Spread Factor']
)

plt.figure(figsize=(8, 6))
sns.heatmap(loadings_fa, annot=True, cmap='coolwarm', center=0)
plt.title("Factor Analysis: Latent Relationship of Features")
plt.show()
```

**Dryness Factor:**
This factor is strongly influenced by BUI, DC, and DMC, indicating long-term fuel dryness conditions. Higher values reflect accumulated dryness in vegetation, making it highly susceptible to ignition. It represents the foundational condition required for fire occurrence.

**Atmospheric Factor:**
This factor is driven by ISI, temperature, FFMC, and negatively by RH, capturing the effect of current weather conditions. It reflects how atmospheric conditions like heat and low humidity contribute to fire ignition and intensity. It plays a key role in determining how easily fires can start and develop.

**Spread Factor**:
This factor is associated with temperature, FFMC, wind speed, and rainfall, representing conditions that influence fire propagation. It highlights how environmental factors like dry fuel, wind, and lack of rain affect the spread of fire. It primarily governs how rapidly and extensively a fire can move once ignited.


```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(X_scaled)
eigenvalues = pca.explained_variance_
```

```python
pca_summary = pd.DataFrame({
    'Component': range(1, len(eigenvalues) + 1),
    'Eigenvalue': eigenvalues,
    'Proportion of Variance': pca.explained_variance_ratio_,
    'Cumulative Variance': np.cumsum(pca.explained_variance_ratio_)
})
print("PCA Eigenvalue Summary:")
display(pca_summary.round(4))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(eigenvalues) + 1), eigenvalues, marker='o', linestyle='-', color='b')
plt.axhline(y=1.0, color='r', linestyle='--', label='Kaiser Threshold (Eigenvalue = 1)')
plt.title("Scree Plot of Eigenvalues", fontsize=14)
plt.xlabel("Component Number", fontsize=12)
plt.ylabel("Eigenvalue", fontsize=12)
plt.xticks(range(1, len(eigenvalues) + 1))
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
pca_3 = PCA(n_components=3)
X_pca = pca_3.fit_transform(X_scaled)

loadings = pd.DataFrame(
    pca_3.components_.T,
    columns=['PC1', 'PC2', 'PC3'],
    index=X.columns
)
print("Eigenvectors (Component Weights):")
print(loadings.round(3))
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(loadings, annot=True, cmap='coolwarm', center=0, fmt=".2f")
plt.title("Principal Component Loadings (Eigenvectors)")
plt.show()
```

**PC1 (The "Fire Danger / Dryness" Index):** This component has strong positive loadings for the FWI indices (FWI, BUI, DMC, ISI) and Temperature, while having a strong negative loading for Relative Humidity (RH). It acts as an overall summary of how hot, dry, and dangerous the environment is.
                                                                                
**PC2 (The "Weather Shift" Index (Wind & Moisture Variability):** This component is dominated by high Wind Speed, Humidity, and Drought Code (DC), but inversely related to Temperature. Instead of measuring pure heat-driven fire risk, it captures dynamic atmospheric changes—such as cool, windy, and humid weather systems moving over historically dry land.

**PC3 (The "Precipitation vs Fire Spread Trade-off"):** This component is characterized by high Rain, Wind Speed, and Spread Index (ISI), but negatively associated with Humidity and DC. It represents acute, windy rainstorms, capturing the trade-off between rain suppressing fires and wind spreading them. The negative humidity loading importantly distinguishes sudden rainfall from general atmospheric moisture.

**Observation: The Trade-off Between Accuracy and Stability**

* **Model 1 (Raw Data)**: Achieved an accuracy of ~97.3%. However, because we proved earlier that this data suffers from severe multicollinearity, this model is mathematically unstable. It is relying on redundant information (like feeding it FWI, BUI, and DMC all at once), which makes its internal coefficients untrustworthy and prone to overfitting.

* **Model 2 (PCA Data)**: Achieved a strong accuracy of ~89.0% using 70% fewer features (only 3 components instead of 10). Furthermore, it trained roughly 4 times faster (0.013s vs 0.058s).


By utilizing Principal Component Analysis, we successfully compressed 10 highly correlated environmental variables into 3 independent, physically meaningful dimensions (Heat/Dryness, Wind Shifts, and Rainstorms).

Final Conclusion: While the raw model achieved slightly higher raw accuracy, it did so using a mathematically flawed, multicollinear dataset. Our PCA-transformed model retained 89% predictive accuracy while utilizing drastically less data and completely eliminating the multicollinearity. We have successfully traded a small amount of redundant accuracy for massive gains in mathematical stability, computational efficiency, and physical interpretability.
